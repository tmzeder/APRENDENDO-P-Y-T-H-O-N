```python
def amostragem_aleatoria_simples(dataset, amostras): # a funçao que retorna o data frame completo onde passamos a base de dados e a amostra
    return dataset.sample(n = amostras, random_state = 1) #o valor de n será igual as amostras
```

```python
df_amostra_aleatoria_simples = amostragem_aleatoria_simples(dataset, 100) #dataset é a base completa e 100 o numero de amostras
df_amostra_aleatoria_simples.shape
```

Amostragem Sistemática

```python
dataset.shape #observamos o tamanho dos registros
```

```python
len(dataset) // 100 # retorna o tamanho do dataset e fazemos a divisão por 100
```

```python
random.seed(1) # configuramos o valor 1 para retornar o mesmo valor sempre - útil caso precise replicar o experimento sem mudar o resultado
random.randint(0, 325) # seleciona um numero randomico entre 0 e 325, a cada vez que executar ele retorna um numero aleatório neste intervalo
```

```python
np.arange(68, len(dataset), step = 325) #função para fazer geração dos numeros, indique o numero que deseja começar  e o numero final da base de dados , no saco o len(dataset) e por fim passamos o outro parametro "STEP" que de quantos em quantos serão retornados os valores
```

```python
def amostragem_sistematica(dataset, amostras): #função pra colocar o código em um só local
    intervalo = len(dataset) // amostras
    random.seed(1) # sempre ter o mesmo valor
    inicio = random.randint(0, intervalo) #buscar o valor inicial - de 0 até o intervalo ( no python os indices começam com zero)
    indices = np.arange(inicio, len(dataset), step = intervalo) #variavel que armazena os indices
    amostra_sistematica = dataset.iloc[indices] #variavel recebe o dataset com o iloc que você vai passar quais indices quer selecionar 
    return amostra_sistematica #retorna 
```

```python
df_amostra_sistematica = amostragem_sistematica(dataset, 100)#recebe a  função passando o dataset e a quantidade de elementos que você quer buscar
df_amostra_sistematica.shape #certifica que serão selecionados corretamente os elementos da amostra
```

```python
def amostragem_agrupamento(dataset, numero_grupos):
    intervalo = len(dataset) / numero_grupos
    grupos = []
    id_grupo = 0
    contagem = 0

    for _ in dataset.iterrows():
        grupos.append(id_grupo)
        contagem += 1
        if contagem > intervalo:
            contagem = 0
            id_grupo += 1

    dataset['grupo'] = grupos
    random.seed(1)
    grupo_selecionado  = random.randint(0, numero_grupos)
    return dataset[dataset['grupo'] == grupo_selecionado]
```
